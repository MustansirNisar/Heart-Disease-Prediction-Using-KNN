### What is KNN?
K-Nearest Neighbors is a simple, yet powerful supervised learning algorithm used for both classification and regression tasks. It's called a "lazy learning" algorithm because it doesn't build an explicit model during training—instead, it stores all training data and makes predictions based on the similarity of new data points to existing ones.

![KNN Concept Visualization](https://miro.medium.com/v2/resize:fit:1400/0*34SajbTO2C5Lvigs.png)

### Key Concepts
- **K**: The number of nearest neighbors to consider
- **Distance Metric**: How we measure similarity between data points
- **Lazy Learning**: No training phase, all computation happens during prediction
- **Instance-Based**: Makes decisions based on stored training instances

---

## 4. How KNN Works

### 4.1 The Algorithm Steps

1. **Store Training Data**: Keep all training examples in memory
2. **Choose K**: Select the number of neighbors to consider
3. **Calculate Distance**: For a new data point, compute distance to all training points
4. **Find K Nearest**: Identify the K closest training examples
5. **Make Prediction**:
   - **Classification**: Vote among K neighbors (majority class)
   - **Regression**: Average the values of K neighbors

### 4.2 Visual Representation

![KNN Classification Example](https://www.ejable.com/wp-content/uploads/2023/11/7-b.webp)

```
Classification Example (K=3):
    
    Classes: ● (Class A), ■ (Class B), ▲ (Class C)
    New Point: ★
    
    Distance Calculation:
    ★ ----2.1---- ●
    ★ ----1.8---- ■
    ★ ----2.5---- ●
    ★ ----1.5---- ■
    ★ ----3.2---- ▲
    ★ ----1.9---- ■
    
    3 Nearest Neighbors: ■, ■, ■
    Prediction: Class B (majority vote)
```

### 4.3 Distance Metrics

#### Euclidean Distance (Most Common)
```
d(p,q) = √[(p₁-q₁)² + (p₂-q₂)² + ... + (pₙ-qₙ)²]
```
**Use Case**: Continuous variables, geometric problems

#### Manhattan Distance
```
d(p,q) = |p₁-q₁| + |p₂-q₂| + ... + |pₙ-qₙ|
```
**Use Case**: Grid-like data, taxi-cab distance

#### Minkowski Distance
```
d(p,q) = (Σ|pᵢ-qᵢ|ᵖ)^(1/p)
```
**Use Case**: Generalization of Euclidean (p=2) and Manhattan (p=1)

#### Cosine Distance
```
d(p,q) = 1 - (p·q)/(|p||q|)
```
**Use Case**: Text mining, high-dimensional sparse data

## 5. Choosing the Right K Value

### Impact of K Value

| K Value | Characteristics | Pros | Cons |
|---------|----------------|------|------|
| K = 1 | Uses only nearest neighbor | Simple, fast | Sensitive to noise |
| Small K | More sensitive to local patterns | Captures local structure | Overfitting risk |
| Large K | More global decision boundary | Stable, less noise | May miss local patterns |
| K = n | Uses all training data | Very stable | Loses local information |

### Guidelines for Choosing K
- **Odd numbers**: Prevents ties in binary classification
- **Cross-validation**: Test different K values
- **Rule of thumb**: K = √n (where n is number of training samples)
- **Domain knowledge**: Consider problem-specific requirements

# K-Nearest Neighbors (KNN) Algorithm

## Introduction

K-Nearest Neighbors (KNN) is a supervised machine learning algorithm primarily used for classification tasks, though it can also handle regression problems. The algorithm operates on a simple yet powerful principle: it predicts the class or value of a new data point by examining the 'k' closest data points (neighbors) in the training dataset.

### Key Characteristics

- **Non-parametric**: Makes no assumptions about the underlying data distribution
- **Instance-based learning**: Uses the actual training instances to make predictions
- **Lazy learner**: Does not build an explicit model during training; instead, it stores the entire dataset and performs computations at prediction time

## How KNN Works

KNN classifies new data points based on similarity and proximity. The algorithm assigns a category to a new point by looking at the majority class among its nearest neighbors.

### Example Classification Process

Consider a scenario where we have:
- **Red diamonds**: Category 1
- **Blue squares**: Category 2
- **New data point**: Unknown category

The algorithm examines the closest neighbors to the new point. If the majority of these neighbors belong to Category 2 (blue squares), then KNN predicts that the new point also belongs to Category 2.

*[Image placeholder: KNN classification example showing red diamonds, blue squares, and a new point with circled neighbors]*

## Understanding 'K' in K-Nearest Neighbors

The parameter 'k' represents the number of nearest neighbors the algorithm considers when making a prediction.

### Simple Analogy

Imagine you're trying to identify an unknown fruit based on its shape and size:
- If **k = 3**, the algorithm examines the 3 closest known fruits
- If 2 out of these 3 fruits are apples and 1 is a banana, the algorithm predicts the unknown fruit is an apple (majority vote)

## Selecting the Optimal Value of K

Choosing the right value of k is crucial for algorithm performance and depends on several factors:

### Considerations for K Selection

- **Noisy data**: Larger k values provide more stable predictions by reducing the impact of outliers
- **Too large k**: May lead to underfitting, causing the model to miss important patterns
- **Data characteristics**: The optimal k should be chosen based on your specific dataset

### Statistical Methods for K Selection

#### 1. Cross-Validation
- Divide the dataset into k parts (k-fold cross-validation)
- Train the model on some parts and test on the remaining ones
- Repeat this process for each part
- Select the k value that yields the highest average accuracy

#### 2. Elbow Method
- Plot a graph showing error rate or accuracy for different k values
- Initially, error decreases as k increases
- Identify the "elbow point" where the error rate stops decreasing rapidly
- This point typically represents the optimal k value

*[Image placeholder: Elbow method graph showing error rate vs k values]*

#### 3. Odd Values for K
- Use odd numbers for k, especially in classification problems
- Helps avoid ties when determining the majority class among neighbors

## Distance Metrics in KNN

KNN relies on distance metrics to identify the nearest neighbors. Here are the most commonly used metrics:

### 1. Euclidean Distance

The straight-line distance between two points in space - the shortest path between them.

**Formula:**
```
distance(x, Xi) = √[Σ(xj - Xij)²]
```

### 2. Manhattan Distance

Also known as "taxicab distance" - the total distance traveled when you can only move along horizontal and vertical lines (like city streets).

**Formula:**
```
d(x,y) = Σ|xi - yi|
```

### 3. Minkowski Distance

A generalized distance metric that includes both Euclidean and Manhattan distances as special cases.

**Formula:**
```
d(x,y) = (Σ|xi - yi|^p)^(1/p)
```

- When **p = 2**: Equivalent to Euclidean distance
- When **p = 1**: Equivalent to Manhattan distance

## KNN Algorithm Workflow

### Step 1: Select Optimal K Value
Determine the number of nearest neighbors to consider for prediction.

### Step 2: Calculate Distances
Compute the distance between the target point and all training data points using a chosen distance metric (typically Euclidean distance).

### Step 3: Identify Nearest Neighbors
Select the k data points with the smallest distances to the target point.

### Step 4: Make Prediction

#### For Classification:
- Use **majority voting** among the k nearest neighbors
- Assign the class that appears most frequently among the neighbors

#### For Regression:
- Calculate the **average** of the values of the k nearest neighbors
- Use this average as the predicted value

## Advantages and Disadvantages

### Advantages
- Simple to understand and implement
- No assumptions about data distribution
- Effective for both classification and regression
- Naturally handles multi-class problems

### Disadvantages
- Computationally expensive for large datasets
- Sensitive to irrelevant features
- Performance depends heavily on the choice of k
- Sensitive to the scale of features (normalization required)

## Implementation Considerations

- **Feature scaling**: Normalize or standardize features to ensure all dimensions contribute equally to distance calculations
- **Curse of dimensionality**: Performance may degrade in high-dimensional spaces
- **Memory requirements**: Stores the entire training dataset
- **Computational complexity**: O(n) for each prediction, where n is the training set size

````python
import pandas as pd
import numpy as np
````

````python
# Load the CSV file into a DataFrame
df = pd.read_csv('heart.csv')

# View the first five rows
df.head()
# df['ST_Slope'].unique()
# df['RestingECG'].unique()
````

````python
# Check shape, column names, data types
df.info()

# Check for missing values
print(df.isnull().sum())
````

````python
df['HeartDisease'].value_counts()
````

````python
df = pd.get_dummies(df, drop_first=True)
df.head()
````

````python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
````

````python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
````

````python
y_pred = knn.predict(X_test)
````

````python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Detailed classification metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
````

````python
import matplotlib.pyplot as plt

error_rates = []
k_range = range(1, 21)

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_k = knn.predict(X_test)
    error = 1 - accuracy_score(y_test, y_pred_k)
    error_rates.append(error)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(k_range, error_rates, marker='o', linestyle='--', color='b')
plt.title('Error Rate vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate (1 - Accuracy)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
````

````python
from sklearn.model_selection import cross_val_score
k_range = range(1, 21)
cv_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    # 10-fold cross-validation on the entire (scaled) dataset
    scores = cross_val_score(knn, X_scaled, y, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
plt.figure(figsize=(10, 5))
plt.plot(k_range, cv_scores, marker='o', linestyle='--', color='green')
plt.title('Cross-Validation Accuracy vs K')
plt.xlabel('K')
plt.ylabel('Cross-Validated Accuracy')
plt.xticks(k_range)
plt.grid(True)
plt.show()
````
